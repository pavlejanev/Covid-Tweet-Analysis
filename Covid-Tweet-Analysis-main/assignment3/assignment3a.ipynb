```python
' Bulk of program in which we will be running everything 50 times to simulate an average for incorrect, correct, and mislabled predictions '
from sklearn.naive_bayes import GaussianNB
count = 1 
total_average_correct = 0
total_average_incorrect = 0

while count <= 50:

    X_train,X_test,y_train,y_test=train_test_split(x_set,y_set,test_size=0.4,random_state=20) # Creating training and test sets 
    nb_gauss_coimbra=GaussianNB()
    nb_gauss_coimbra  #Initializing Gaussian NB
    X_train,X_test,y_train,y_test=train_test_split(x_set,y_set,test_size=0.3,random_state=20)
    nb_gauss_coimbra=GaussianNB()
    nb_gauss_coimbra
    
    nb_gauss_coimbra.fit(X_train,y_train)
    y_pred = nb_gauss_coimbra.predict(X_test)
    nb_gauss_coimbra.score(X_train,y_train)
    nb_gauss_coimbra.score(X_test,y_test)
    
    print("Run: %s" % count)
    print("Accuracy: %s" % accuracy_score(y_pred,y_test))
    probability = nb_gauss_coimbra.predict_proba(X_test).max(axis=1)
    high_confidence = probability > 0.9
   

    average_correct = (probability[y_test == y_pred].sum()) /  (y_test ==y_pred).sum()
    average_incorrect = (probability[y_test != y_pred].sum()) /  (y_test !=y_pred).sum()
    
    total_average_correct += average_correct
    total_average_incorrect = average_incorrect

    print("Average chance of a correct prediction : %s" % average_correct)
    print("Average chance of an incorrect prediction : %s" % average_incorrect)\
    
    print("Number of mislabeled points out of a total",X_test.shape[0],"points:",(y_test != y_pred).sum())
    
    count += 1 
    

```
